```python

import ares.Lib.Ares as Ares

df = aresObj.df({
    'dates': ['2019-01-01', '2019-01-01', '2019-01-10', '2019-01-11'],
    'category': ["ABCDEF", "ABCDEF", "DFCGHI", "GHIKLM"],
    'series': [3, 3, 4, 6],
    'series2': [5, 5, 1, 9],
    'series3': [3, 7, 3, 3]})

aresObj = Ares.ReportAPI()
aresObj.checkbox(df, title="Test", dfColumn='category', allSelected=True, globalFilter=True, htmlCode='MCheck', icon="fas fa-circle")
aresObj.display()
```
