```python
recordSet = {
    'dates': ['2019-01-01', '2019-01-01', '2019-01-10', '2019-01-11'],
    'category': ["ABCDEF", "ABCDEF", "DFCGHI", "GHIKLM"],
    'series': [3, 3, 4, 6],
    'series2': [5, 5, 1, 9],
    'series3': [3, 7, 3, 3]}

import ares.Lib.Ares as Ares

aresObj = Ares.ReportAPI()

df = aresObj.df(recordSet)
table = aresObj.table(df)
c = aresObj.chart('bar', df, seriesNames=['series', 'series2'], xAxis='dates')

t = aresObj.tabs([], selectedTab='My Chart')
t + (table, 'My Table')
t + (c, 'My Chart')

aresObj.display()
```
