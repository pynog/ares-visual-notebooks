```python
recordSet = {
    'dates': ['2019-01-01', '2019-01-01', '2019-01-10', '2019-01-11'],
    'category': ["ABCDEF", "ABCDEF", "DFCGHI", "GHIKLM"],
    'series': [300000, 300000, 400000, 600000],
    'series2': [-500000, 500000, 100000, 900000],
    'series3': [3.0167, 7, 3, 3]}
```

```python
import ares.Lib.Ares as Ares

aresObj = Ares.ReportAPI()
data = aresObj.df(recordSet)
aresObj.checkbox(data, dfColumn='dates')
aresObj.display()
```

```python
import ares.Lib.Ares as Ares

aresObj = Ares.ReportAPI()
data = aresObj.df(recordSet)
c1 = aresObj.checkbox(data, dfColumn='dates')
c2 = aresObj.checkbox(data, dfColumn='category')
aresObj.grid([c1, c2], colsDim=[2, 2])
aresObj.display()
```

```python
import ares.Lib.Ares as Ares

aresObj = Ares.ReportAPI()
data = aresObj.df(recordSet)
c1 = aresObj.checkbox(data, dfColumn='dates')
c2 = aresObj.checkbox(data, dfColumn='category').click(aresObj.jsConsole())
aresObj.grid([c1, c2], colsDim=[2, 2])
aresObj.display()
```

```python
import ares.Lib.Ares as Ares

aresObj = Ares.ReportAPI()
data = aresObj.df(recordSet)
c1 = aresObj.checkbox(data, dfColumn='dates')
c2 = aresObj.checkbox(data, dfColumn='category')
c1.click(
    [
      aresObj.jsConsole(),
      c2.jsDisable([]),
      aresObj.jsRule('2019-01-01', c2.jsDisable(["ABCDEF"], reset=False), jsCondition="%(srcVal)s.indexOf(%(pyVal)s) > -1", srcVal=c1.val),
      aresObj.jsRule('2019-01-10', c2.jsDisable(["DFCGHI", "GHIKLM"], reset=False), jsCondition="%(srcVal)s.indexOf(%(pyVal)s) > -1", srcVal=c1.val)
    ])
aresObj.grid([c1, c2], colsDim=[2, 2])
aresObj.display()
```

```python
import ares.Lib.Ares as Ares

aresObj = Ares.ReportAPI()
data = aresObj.df(recordSet)
c2 = aresObj.checkbox(data, dfColumn='category')
aresObj.button("Remove DFCGHI").click(c2.jsRemove(["DFCGHI"]))
aresObj.display()
```

```python
import ares.Lib.Ares as Ares

aresObj = Ares.ReportAPI()
data = aresObj.df(recordSet)
c2 = aresObj.checkbox(data, dfColumn='category')
aresObj.button("Add Category").click(c2.jsAdd([{"name": 'Test'}]))
aresObj.display()
```
