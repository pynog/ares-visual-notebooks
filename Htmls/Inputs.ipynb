```python
import ares.Lib.Ares as Ares

aresObj = Ares.ReportAPI()
b = aresObj.button("Save as", align=None).isDisable()
i = aresObj.input(placeholder="Config Name", widthUnit="px").cssInput({"text-align": "left"}).keyup([
    b.jsDisable(True, isPyData=True),
    aresObj.jsRule("", b.jsDisable(False, isPyData=True), rule="!=")
])
b.click([
    i.jsClear(),
    b.jsDisable(True, isPyData=True),
])
aresObj.div([b, i])
aresObj.display()
```
