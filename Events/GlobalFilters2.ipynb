```python
import random

series = 5
steps = 50
threshold = 34

# Generate the data
# Index is added automatically on demand from the dataframe
data, cols, category, orderCols = {}, [], [], ['_index', 'category', 'product']
cats = ['A', 'B', 'C', 'D']
for i in range(series):
    data["Series%s" % i] = [random.uniform(-100, 100) for _ in range(steps)]
    cols.append("Series%s" % i)
    orderCols.append("Series%s" % i)
data['category'] = [cats[int(random.uniform(-100, 100) % 3)] for _ in range(steps)]
data['product'] = [int(random.uniform(-100, 100) % 3) for _ in range(steps)]

import ares.Lib.Ares as Ares

aresObj = Ares.ReportAPI()

# Dashboard
df = aresObj.df(data)
p = aresObj.chart('polar', df, seriesNames=['Series0'], xAxis='category', globalFilter=True, htmlCode='pieChart')
b = aresObj.chart('radar', df, seriesNames=['Series0', 'Series1', 'Series2'], otherDims=['Series3'], globalFilter=True, htmlCode='barChart')

t = aresObj.table(df, htmlCode='tableData', tableOptions={'pageLength': 10})
aresObj.grid([aresObj.col([p, b]), t], colsDim=[6, 6])  # Put the charts on the same row

# Events
p.click([p.jsGenerate(None), b.jsGenerate(None), t.jsGenerate(None)])
b.click([p.jsGenerate(None), b.jsGenerate(None), t.jsGenerate(None)])

aresObj.display()
```
