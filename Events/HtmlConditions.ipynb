```python
import ares.Lib.Ares as Ares

aresObj = Ares.ReportAPI()
s1 = aresObj.checkbox(["1", "2", "3"])
s2 = aresObj.checkbox([])
i = aresObj.input()

s1.click([
    aresObj.jsRule("2", s2.jsGenerate([{"value": "G", 'name': "Label"}, {"value": "H"}], isPyData=True), srcVal='data.event_label'),
    aresObj.jsRule("3", s2.jsGenerate([], isPyData=True), srcVal='data.event_label'),
    aresObj.jsRule("1", [
      i.jsGenerate("Value 1 selected", isPyData=True),
      s2.jsGenerate([{"value": "W"}, {"value": "z"}], isPyData=True)
    ], srcVal='data.event_label'),
  ])

s2.click([
    aresObj.jsRule("z", [
      i.jsGenerate("Value z selected", isPyData=True)
    ], srcVal='data.event_label')
  ])
aresObj.display()
```
