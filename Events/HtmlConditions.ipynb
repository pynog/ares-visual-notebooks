```python
import ares.Lib.Ares as Ares

aresObj = Ares.ReportAPI()
s1 = aresObj.select(["1", "2", "3"], search=False)
s2 = aresObj.select([], search=False)
i = aresObj.input()

s1.change([
    aresObj.jsRule("2", s2.jsGenerate([{"value": "G", 'name': "Label"}, {"value": "H"}], isPyData=True)),
    aresObj.jsRule("3", s2.jsGenerate([], isPyData=True)),
    aresObj.jsRule("1", [
        i.jsGenerate("Value 1 selected", isPyData=True),
        s2.jsGenerate([{"value": "W"}, {"value": "z"}], isPyData=True)
    ]),
])
s2.change([
    aresObj.jsRule("z", [
        i.jsGenerate("Value z selected", isPyData=True)
    ])
])

aresObj.newline(6)
aresObj.display()
```
