```python
import ares.Lib.Ares as Ares

aresObj = Ares.ReportAPI()
i = aresObj.input("")

# Use the browser storage available in the tab (HTML5 feature)
aresObj.button("Load").click(aresObj.jsAddToLocalStorage("MyData", i.val))
aresObj.button("Display").click(aresObj.jsAlert(aresObj.jsGetFromLocalStorage("MyData")))
aresObj.display()
```

```python
import ares.Lib.Ares as Ares

aresObj = Ares.ReportAPI()
# Result visible in the debug window of the browser (F12)
interv = aresObj.jsConsole("Run Function", isPyData=True)
intId = aresObj.jsInterval(interv, 500)
aresObj.button("Stop").click(aresObj.jsClearInterval(intId))
aresObj.display()
```
