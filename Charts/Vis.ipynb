```python
import random
import time

step = 10
series = 5
xAxisCat = ['A', 'B', 'C', 'D', 'E']

def strTimeProp(start, end, format, prop):
    """Get a time at a proportion of a range of two formatted times.

    start and end should be strings specifying times formated in the
    given format (strftime-style), giving an interval [start, end].
    prop specifies how a proportion of the interval to be taken after
    start.  The returned time will be in the specified format.
    """
    stime = time.mktime(time.strptime(start, format))
    etime = time.mktime(time.strptime(end, format))
    ptime = stime + prop * (etime - stime)
    return time.strftime(format, time.localtime(ptime))

def randomDate(start, end, prop):
    return strTimeProp(start, end, '%Y-%m-%d', prop)

# Create the data set used in the chart objects
recordSet = {'categories': [xAxisCat[int(random.random()*100) % len(xAxisCat)] for i in range(step)],
             'dates': [randomDate('2018-01-01', '2018-12-31', random.random())for i in range(step)]}
for i in range(series):
    recordSet["col%s" % i] = [random.random()*100 for _ in range(step)]
# Create an AReS DataFrame from the python dictionary

import ares.Lib.Ares as Ares
aresObj = Ares.ReportAPI()
df = aresObj.df(recordSet)

aresObj.chart('bubble', df, title='Bubble 3D', seriesNames=['col2', 'col1'], xAxis='col3', otherDims=['col0'], chartFamily='Vis')

aresObj.display()
```
