```python
import ares.Lib.Ares as Ares

aresObj = Ares.ReportAPI()

df1 = aresObj.df(
  [
      {'id': 1, 'label': 'A', 'group': 'user', 'shape': 'star', 'value': 0.2, 'content': 'item 1', 'start': '2014-04-20', 'end': '2014-04-26'},
      {'id': 2, 'label': 'B', 'group': 'user', 'shape': 'circle', 'value': 0.6, 'content': 'item 2', 'start': '2014-05-14', 'end': '2014-05-18'},
      {'id': 3, 'label': 'C', 'group': 'user', 'shape': 'circle', 'type': 'point', 'content': 'item 3', 'start': '2014-04-15', 'end': '2014-05-18'}
    ]
  )
c = aresObj.chart('network', df1, title='Network Chart', chartFamily='Vis')
c.edges([
    {'from': 1, 'to': 3, 'arrows': 'to'},
    {'from': 1, 'to': 2, 'dashes': True},
    {'from': 2, 'to': 2},
    {'from': 2, 'to': 3, 'value': 10},
])
aresObj.display()
```
