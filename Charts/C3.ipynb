```python
import random

step = 10
series = 5
xAxisCat = ['A', 'B', 'C', 'D', 'E']
# Create the data set used in the chart objects
recordSet = {'categories': [xAxisCat[int(random.random()*100) % len(xAxisCat)] for i in range(step)]}
for i in range(series):
    recordSet["col%s" % i] = [random.random()*100 for _ in range(step)]
# Create an AReS DataFrame from the python dictionary
```

```python
import ares.Lib.Ares as Ares

aresObj = Ares.ReportAPI()
df = aresObj.df(recordSet)
l = aresObj.chart('line', df, title='Line Chart', seriesNames=['col1', 'col2'], xAxis='categories', chartFamily='C3')
aresObj.display()
```

```python
import ares.Lib.Ares as Ares

aresObj = Ares.ReportAPI()
df = aresObj.df(recordSet)
a = aresObj.chart('area', df, seriesNames=['col0', 'col3'], xAxis='categories', chartFamily='C3', chartOptions={'legend': {'show': False}})
aresObj.display()
```

```python
import ares.Lib.Ares as Ares

aresObj = Ares.ReportAPI()
df = aresObj.df(recordSet)
p = aresObj.chart('pie', df, seriesNames=['col0'], xAxis='categories', chartFamily='C3')
d = aresObj.chart('donut', df, seriesNames=['col3'], xAxis='categories', chartFamily='C3', chartOptions={'legend': {'show': False}})
d.addAttr('title', 'Test', 'donut')    
aresObj.display()
```
