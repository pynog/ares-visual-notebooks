```python
import ares.Lib.Ares as Ares
import random

step = 10
series = 5
xAxisCat = ['A', 'B', 'C', 'D', 'E']

aresObj = Ares.ReportAPI()

recordSet = {'categories': [xAxisCat[int(random.random()*100) % len(xAxisCat)] for i in range(step)]}
for i in range(series):
    recordSet["col%s" % i] = [random.random()*100 for _ in range(step)]
# Create an AReS DataFrame from the python dictionary
df1 = aresObj.df(recordSet)

c = aresObj.chart('pie', df1, title='Pie Chart', seriesNames=['col2', 'col1'], xAxis='categories')
l = aresObj.chart('area-end', df1, title='Area Chart', seriesNames=['col3', 'col4'], xAxis='categories')
b = aresObj.chart('bar', df1, seriesNames=['col0', 'col1'], xAxis='categories', htmlCode='ChartBar')
p = aresObj.chart('polar', df1, seriesNames=['col0'], xAxis='categories', htmlCode='ChartPolar')
aresObj.grid([l, b, c], colsDim=[4, 4, 4])

aresObj.display()
```

```python
import ares.Lib.Ares as Ares
import random

step = 10
series = 5
xAxisCat = ['A', 'B', 'C', 'D', 'E']

aresObj = Ares.ReportAPI()

recordSet = {'categories': [xAxisCat[int(random.random()*100) % len(xAxisCat)] for i in range(step)]}
for i in range(series):
    recordSet["col%s" % i] = [random.random()*100 for _ in range(step)]
# Create an AReS DataFrame from the python dictionary
df1 = aresObj.df(recordSet)

aresObj.chart('bubble', df1, seriesNames=['col%s' % i for i in range(series)], htmlCode='ChartBubble')
aresObj.chart('radar', df1, seriesNames=['col2', 'col3', 'col4', 'col0'], xAxis='categories', htmlCode='ChartRadar')
aresObj.chart('donut', df1, seriesNames=['col3'], xAxis='categories', globalFilter=True, htmlCode='ChartDonut')


aresObj.display()
```
