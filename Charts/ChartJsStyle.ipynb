```python
import random 

step = 20
series = 6

recordSet = {'x': range(step)}
for i in range(series):
    recordSet['col%s' % i] = [random.random() for i in range(step)]
```

```python
import random
import ares.Lib.Ares as Ares

aresObj = Ares.ReportAPI()
df = aresObj.df(recordSet)
c = aresObj.chart('bar', df, seriesNames=["col%s" % i for i in range(4)], xAxis='x')

# Style the chart
c.addSeriesAttr(0, data={'type': 'line', 'fill': False, 'showLine': False, 'backgroundColor': 'green', 'pointStyle': 'rect'})
c.addSeries('scatter', 'col4', [random.random() for i in range(step)], options={'fill': False, 'pointStyle': 'triangle', 'showLine': False}, color='red')
c.addPoints('col6', [random.random() for i in range(step)], options={'pointRadius': 5})

aresObj.display()
```

```python
import random
import ares.Lib.Ares as Ares

aresObj = Ares.ReportAPI()
df = aresObj.df(recordSet)
c = aresObj.chart('bar', df, seriesNames=[], xAxis='x')
c.yFormat('K')

# Style the chart
c.addTriangles('col7', [random.random() * 1000000 for i in range(step)])
aresObj.display()
```
