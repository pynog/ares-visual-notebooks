```python
import ares.Lib.Ares as Ares

aresObj = Ares.ReportAPI()
aresObj.setTheme('dark')
s = aresObj.search()
b = aresObj.optionsbar([
    {"icon": "fas fa-comment-dots", 'jsFnc': aresObj.jsConsole(), 'tooltip': 'Youpi'},
    {"icon": "fas fa-video", 'jsFnc': aresObj.jsConsole()},
    {"icon": "fas fa-bell", 'jsFnc': aresObj.jsConsole()}
  ])
aresObj.grid([s, b], colsDim=[10, 2]).css({'overflow': 'hidden'})

aresObj.comments([])
c1 = aresObj.col([
      aresObj.media("myVideo.flv", path="."),
      aresObj.stars(2, align='right').click([aresObj.jsConsole()]),
      aresObj.title("My New Video", level=3)]).css({'padding': '5px'})

c2 = aresObj.col([
    aresObj.title("Youpi", level=4),
    aresObj.switch({'on': 'A', 'off': 'N'}, label="automatic"),
    aresObj.vignet({'title': 'Python', 'number': 100, 'text': 'Ares new Python web framework', 'color': 'green',
                    'url': 'https://www.python.org/',
                    'icon': 'fab fa-python', 'tooltip': 'Python Fondation', 'urlTitle': 'WebSite'}),
    aresObj.textbubble({"url": "", "title": "Python", 'value': 100, 'color': 'green'}, size=40)

  ])

aresObj.grid([c1, c2], colsDim=[9, 3])
aresObj.display()
```
