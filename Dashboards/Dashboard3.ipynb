```python
recordSet = {
    'dates': ['2019-01-01', '2019-01-01', '2019-01-10', '2019-01-11'],
    'category': ["ABCDEF", "ABCDEF", "DFCGHI", "GHIKLM"],
    'category2': ["wcec", "ece", "ece", "ewc"],
    'series': [3, 3, 4, 6],
    'series2': [5, 5, 1, 9],
    'series3': [3, 7, 3, 3]}
```

```python
import ares.Lib.Ares as Ares

aresObj = Ares.ReportAPI()
data = aresObj.df(recordSet)
c1 = aresObj.chart("hbar", data, seriesNames=['series'], xAxis='category', globalFilter=True, htmlCode="chart1")
c2 = aresObj.chart('donut', data, seriesNames=['series2'], title="series2", xAxis='category', globalFilter=True, htmlCode="chart2")
c3 = aresObj.chart('donut', data, seriesNames=['series3'], title="series3", xAxis='category', globalFilter=True, htmlCode="chart3")
c4 = aresObj.chart('donut', data, seriesNames=['series'], title="series", xAxis='category2', globalFilter=True, htmlCode="chart4")
aresObj.grid([c2, c3, c4], colsDim=[4, 4, 4])

c1.click([c2.jsGenerate(None), c3.jsGenerate(None), c4.jsGenerate(None)])
c2.click([c1.jsGenerate(None), c3.jsGenerate(None), c4.jsGenerate(None)])
c3.click([c1.jsGenerate(None), c2.jsGenerate(None), c4.jsGenerate(None)])
c4.click([c1.jsGenerate(None), c3.jsGenerate(None), c2.jsGenerate(None)])
aresObj.display()
```
