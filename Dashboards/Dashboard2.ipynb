```python
import random 

step = 20
series = 6

recordSet = {'x': range(step)}
for i in range(series):
    recordSet['col%s' % i] = [random.random() for i in range(step)]
```

```python
import ares.Lib.Ares as Ares

aresObj = Ares.ReportAPI()
df = aresObj.df(recordSet)

c = aresObj.col([])
c + aresObj.chart('line', df, title="Categories", seriesNames=["col%s" % i for i in range(series)], xAxis='x')
c + aresObj.table(df, tableOptions={'pageLength': 10})

aresObj.grid([
    c, aresObj.chart('bar', df, seriesNames=['col0'], xAxis='x', height=700)
], colsDim=[8, 4])

aresObj.display()
```
