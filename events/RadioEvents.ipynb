```python
import random

series = 5
steps = 50
threshold = 34
chart = None #'Billboard'

# Generate the data
# Index is added automatically on demand from the dataframe
data, cols, category, orderCols = {}, [], [], ['_index', 'category', 'product']
prds = ['A', 'B', 'C']
for i in range(series):
    data["Series%s" % i] = [random.uniform(-100, 100) for _ in range(steps)]
    cols.append("Series%s" % i)
    orderCols.append("Series%s" % i)
data['category'] = ['Above' if random.uniform(-100, 100) > threshold else 'Below' for _ in range(steps)]
data['product'] = [prds[int(random.uniform(-100, 100) % 3)] for _ in range(steps)]

import ares.Lib.Ares as Ares

aresObj = Ares.ReportAPI()
# Dashboard
df = aresObj.df(data)

s = aresObj.select(df, htmlCode="cat", dfColumn='category', allSelected=True, globalFilter=True)
r = aresObj.radio(df, htmlCode="product", dfColumn='product', withRemoveButton=True, globalFilter=True)
p = aresObj.chart('pie', df, seriesNames=['Series0'], xAxis='category', chartFamily=chart)
t = aresObj.table(df, htmlCode='tableData', tableOptions={'pageLength': 10})

aresObj.grid([aresObj.col([r, s, p]), t], colsDim=[6, 6])  # Put the charts on the same row

# Events
s.change([p.jsGenerate(None), t.jsGenerate(None)])
r.click([p.jsGenerate(None), t.jsGenerate(None)])

aresObj.display()
```
