```python
import random

series = 5
steps = 50
threshold = 34

# Generate the data
# Index is added automatically on demand from the dataframe
data, cols, category, orderCols = {}, [], [], ['_index', 'category', 'product']
data2 = {}
for i in range(series):
    if i < 4:
      # Not possible to get negative values in a pie chart
      data["Series%s" % i] = [random.uniform(0, 1) for _ in range(steps)]
      data2["Values%s" % i] = [random.uniform(0, 1) for _ in range(steps)]
    else:
      data["Series%s" % i] = [random.uniform(-100, 100) for _ in range(steps)]
      data2["Values%s" % i] = [random.uniform(-100, 100) for _ in range(steps)]
    cols.append("Series%s" % i)
    orderCols.append("Series%s" % i)
data['category'] = ['Above' if random.uniform(-100, 100) > threshold else 'Below' for _ in range(steps)]
data['product'] = [int(random.uniform(-100, 100) % 3) for _ in range(steps)]
data2['product'] = [int(random.uniform(-100, 100) % 3) for _ in range(steps)]

import ares.Lib.Ares as Ares

aresObj = Ares.ReportAPI()
df = aresObj.df(data)
df2 = aresObj.df(data2)

p = aresObj.chart('pie', df, seriesNames=['Series0'], xAxis='category', globalFilter=True, htmlCode='pieChart')
b = aresObj.chart('bubble', df, seriesNames=['Series0', 'Series1', 'Series2'], xAxis='product', otherDims=['Series3'], globalFilter=True, htmlCode='bubbleChart')
aresObj.grid([p, b], colsDim=[6, 6]) # Put the charts on the same row

t = aresObj.table(df, htmlCode='tableData2', tableOptions={'pageLength': 10})

p.click([p.jsGenerate(None), b.jsGenerate(None), t.jsGenerate(None)])
  
aresObj.display()
```
