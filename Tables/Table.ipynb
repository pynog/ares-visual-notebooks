```python
recordSet = {
    'dates': ['2019-01-01', '2019-01-01', '2019-01-10', '2019-01-11'],
    'category': ["ABCDEF", "ABCDEF", "DFCGHI", "GHIKLM"],
    'series': [3, 3, 4, 6],
    'series2': [5, 5, 1, 9],
    'series3': [3, 7, 3, 3]}
```

```python
import ares.Lib.Ares as Ares

aresObj = Ares.ReportAPI()
data = aresObj.df(recordSet)
aresObj.table(data)
aresObj.display()
```

```python
import ares.Lib.Ares as Ares

aresObj = Ares.ReportAPI()
data = aresObj.df(recordSet)
aresObj.table(data).filterHeaders()
aresObj.display()
```

```python
import ares.Lib.Ares as Ares

class CssTableHeader(object):
    __style = [
      {'attr': 'background', 'value': '#59baa6'},
      {'attr': 'text-align', 'value': 'center'},
      {'attr': 'color', 'value': 'white'}]
    childrenTag = 'th'
    
aresObj = Ares.ReportAPI()
data = aresObj.df(recordSet)
aresObj.table(data).setTableCss([CssTableHeader])
aresObj.display()
```

```python
import ares.Lib.Ares as Ares

class CssTablePrevData(object):
    __style = [{'attr': 'background', 'value': '#e2f1ee'}]
    htmlTag = 'td'

aresObj = Ares.ReportAPI()
css = aresObj.addPyCls([CssTablePrevData])
data = aresObj.df(recordSet)
aresObj.table(data, header={
    '_index': {'visible': False},
    'series3': {'class': css[0].classname}})
aresObj.display()
```

```python
dataSet = {'col1': [1, 2, 5], 'col2': [3, None, 8], 'col3': [1, 0, -1],
         'col2.age': [0.2, 0.1, 0], 'col3.age': [0, 0.5, 0],
         'col2.quality': [0.2, 0.1, 0], 'col3.quality': [1, 0.5, 0],
         'col2.intensity': [0.2, 0.6, 0], 'col3.intensity': [1, 0.5, 0]
    }
```

```python
import ares.Lib.Ares as Ares

class CssTableHeader(object):
      __style = [
        {'attr': 'background', 'value': '#59baa6'},
        {'attr': 'border', 'value': '1px solid black'},
        {'attr': 'color', 'value': 'white'}]
    
aresObj = Ares.ReportAPI()
css = aresObj.addPyCls(CssTableHeader)
df = aresObj.df(dataSet)
aresObj.table(df, header={
        'col2': {'title': 'RRR', 'class': css.classname},
        'col3': {'title': 'Signal'},
        '_order': ['col1', 'col2', 'col3']
        },
        cols=['col2', 'col3'], rows=['col1'],
        tableOptions={'system': {'age': True, 'quality': True},
                      'style': {
                          'intensity': {'attr': {'end': 'blue'}},
                          'quality': {'attr': {'end': 'green'}}
                      }}
    )
aresObj.display()
```

```python
import ares.Lib.Ares as Ares

class CssTableOddAllRows(object):
    __style = [{'attr': 'background', 'value': '#e2f1ee'}]
    childrenTag = 'tbody tr:nth-child(odd)'
    hover = [{'attr': 'background', 'value': 'green'}]

aresObj = Ares.ReportAPI()
df = aresObj.df(dataSet)
css = aresObj.addPyCls(CssTableOddAllRows)
aresObj.table(df).addClass(css)
aresObj.display()
```

```python
import ares.Lib.Ares as Ares

aresObj = Ares.ReportAPI()


class CssHeaderLine(object):
    __style = [{'attr': 'background', 'value': '#e2f1ee'}]

class CssHeaderLineRed(object):
    __style = [{'attr': 'color', 'value': 'red'}]

    
    
css = aresObj.addPyCls([CssHeaderLine, CssHeaderLineRed])
data = aresObj.df(recordSet)
tb = aresObj.table(data)
# extraHeaders=None, cssClsStyles=None, cssClsHeader=None
tb.setHeader(extraHeaders=[
    [{'value': 'a', 'title': 'This is a title',  'cols': 2, 'rows': 2}, 
     {'value': 'b', 'cols': 4, 'rows': 1}],
    [{'value': 'd', 'cols': 4, 'rows': 1}]],
    cssClsStyles=[None, css[0].classname],
    cssClsHeader=css[1]
)
aresObj.display()
```

```python
import ares.Lib.Ares as Ares

aresObj = Ares.ReportAPI()
data = aresObj.df(recordSet)
# Sort on the 4th column in the table
aresObj.table(data, tableOptions={'order': [[3, "desc"]]})
aresObj.display()
```

```python
import ares.Lib.Ares as Ares

aresObj = Ares.ReportAPI()
data = aresObj.df(recordSet)
tb = aresObj.table(data)
# Transform only some specific columns
tb.styleColsByNames(['series'], 'number-format', attr={'suffix': '%', 'digits': 2, 'threshold': 4, 
                                                       'factor': 100, 'color': 'red'})
aresObj.display()
```

```python
import ares.Lib.Ares as Ares

aresObj = Ares.ReportAPI()
data = aresObj.df(recordSet)
# Transform all the columns
aresObj.table(data, tableOptions={'style': {'values': {'name': 'number-format', 'attr': {'suffix': '%', 'digits': 2, 'factor': 100}}}})
aresObj.display()
```

```python
recordSet = {
    'dates': ["A1", "A1", "A1", "B1", "C1", "B1", "R1", "E1", "W1"],
    'category': ["A", "B", "C", "D", "A", "B", "R", "E", "W"],
    'series': [8, 3, 4, 0, 7, 8, 0, 67, 0]}
```

```python
import ares.Lib.Ares as Ares

aresObj = Ares.ReportAPI()
data = aresObj.df(recordSet)
c1 = aresObj.chart("bar", data, seriesNames=['series'], xAxis='dates', htmlCode="chart", globalFilter=True)
t = aresObj.table(data)
t.clickRow([
    data.jsFilter("category", jsDataKey="row", jsFnc="function(rec){return JSON.parse(rec).category}"),
    c1.jsGenerate(None)
])
aresObj.display()
```
