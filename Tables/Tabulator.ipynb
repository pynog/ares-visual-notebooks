AReS examples of use of the Javascript library: http://www.tabulator.info/

```python
recordSet = {
    'dates': ['2019-01-01', '2019-01-01', '2019-01-10', '2019-01-11'],
    'category': ["ABCDEF", "ABCDEF", "DFCGHI", "GHIKLM"],
    'series': [3, 3, 4, 6],
    'series2': [5, 5, 1, 9],
    'series3': [3, 7, 3, 3],
    'selected': [True, False, True, True]}
```

```python
import ares.Lib.Ares as Ares

aresObj = Ares.ReportAPI()
data = aresObj.df(recordSet)
aresObj.tabulator(data)
aresObj.display()
```

```python
import ares.Lib.Ares as Ares

aresObj = Ares.ReportAPI()
data = aresObj.df(recordSet)
tb = aresObj.tabulator(data,
        header={'category': {"editor": "select", "editorParams":{"Male":"male", "Female":"female"}},
                'dates': {'editor': 'input'},
                'series': {"editor": 'number'},
                'series2': {"formatter": "buttonCross", "align": "center", "cellClick": aresObj.jsConsole()},
                'selected': {'editor': 'tick', "formatter": "tickCross", 
                             "editorParams":{"tristate": True, "indeterminateValue": "n/a"} }})
aresObj.display()
```

```python
import ares.Lib.Ares as Ares

aresObj = Ares.ReportAPI()
data = aresObj.df(recordSet)
tb = aresObj.tabulator(data)
aresObj.button('clear').click(tb.jsClear())
aresObj.display()
```

```python
import ares.Lib.Ares as Ares

aresObj = Ares.ReportAPI()
data = aresObj.df(recordSet)
tb = aresObj.tabulator(data)
# To change the column dates when series3 is clicked
tb.clickCol(tb.jsUpdateRow("{dates: Today()}"), ['series3'])
aresObj.display()
```

```python
import ares.Lib.Ares as Ares

aresObj = Ares.ReportAPI()
data = aresObj.df(recordSet)
tb2 = aresObj.tabulator(data)
aresObj.button('Add Row').click(tb2.jsAddRow({"category": 'Youpi'}, isPyData=True))
aresObj.display()
```

```python
import ares.Lib.Ares as Ares

aresObj = Ares.ReportAPI()
data = aresObj.df(recordSet)
tb2 = aresObj.tabulator(data).setHeader(extraHeaders={"All Series": ["series", "series2", "series3"]})
aresObj.display()
```

```python
import ares.Lib.Ares as Ares

aresObj = Ares.ReportAPI()
data = aresObj.df(recordSet)
class MyStyle(object):
    __style = [
      {'attr': 'background-color', 'value': 'orange'},
    ]
css = aresObj.addPyCls(MyStyle)
aresObj.tabulator(data, title="title", header={
    "dates": {'cssClass': css.classname},
    "category": {},
    "category2": {}})
aresObj.display()
```

```python
import ares.Lib.Ares as Ares

aresObj = Ares.ReportAPI()
data = aresObj.df(recordSet)
class MyStyle3(object):
    __style = [
      {'attr': 'background-color', 'value': 'orange'},
    ]
    hover = [
      {'attr': 'background-color', 'value': 'green'},
    ]
    parentTag = '.tabulator-row-even'
    
css = aresObj.addPyCls(MyStyle3)
aresObj.tabulator(data, title="title", header={
    "dates": {'cssClass': css.classname},
    "category": {},
    "category2": {}})
aresObj.display()
```

```python
recordSet = {
    'dates': ["A1", "A1", "A1", "B1", "C1", "B1", "R1", "E1", "W1"],
    'category': ["A", "B", "C", "D", "A", "B", "R", "E", "W"],
    'series': [8, 3, 4, 0, 7, 8, 0, 67, 0]}
```

```python
import ares.Lib.Ares as Ares

aresObj = Ares.ReportAPI()
data = aresObj.df(recordSet)
c1 = aresObj.chart("bar", data, seriesNames=['series'], xAxis='dates', htmlCode="chart", globalFilter=True)
t = aresObj.tabulator(data)
t.clickRow([
    data.jsFilter("category", jsDataKey="category"),
    c1.jsGenerate(None)
])
aresObj.display()
```
