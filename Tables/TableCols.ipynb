```python
recordSet = {
    'dates': ['2019-01-01', '2019-01-01', '2019-01-10', '2019-01-11'],
    'category': ["ABCDEF", "ABCDEF", "DFCGHI", "GHIKLM"],
    'series': [3, 3, 4, 6],
    'series2': [5, 5, 1, 9],
    'series4': [1, 0, -1, 1],
    'series3': [3, 7, 3, 3]}
```

```python
import ares.Lib.Ares as Ares

aresObj = Ares.ReportAPI()
data = aresObj.df(recordSet)
aresObj.table(data).styleColsByNames(['series4'], 'url', attr={'link': 'wwww.google.com'}) 
aresObj.display()
```
