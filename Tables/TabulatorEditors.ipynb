AReS examples of use of the Javascript library: http://www.tabulator.info/

```python
recordSet = {
    'dates': ['2019-01-01', '2019-01-01', '2019-01-10', '2019-01-11'],
    'category': ["ABCDEF", "ABCDEF", "DFCGHI", "GHIKLM"],
    'series': [3, 3, 4, 6],
    'series2': [5, 5, 1, 9],
    'series3': [3, 7, 3, 3],
    'selected': [True, False, True, True]}
```

```python
import ares.Lib.Ares as Ares

aresObj = Ares.ReportAPI()
data = aresObj.df(recordSet)
tb = aresObj.tabulator(data, header={
    'series': {'field': 'series', 'title': 'Value', 'formatter': 'number', 'editor': 'input'},
    'dates': {'field': 'dates', 'title': 'Date', 'editor': 'date'},
    'category': {'field': 'category', 'title': 'Category', 'editor' : 'selectConditions',
                 "editorParams": {'key': 'dates', 'values': {"2019-01-01": ["A", "B"]}, 'default': ["r", "u"]}},
  })
aresObj.button("Click").click(aresObj.jsConsole(tb.json))
aresObj.display()
```
